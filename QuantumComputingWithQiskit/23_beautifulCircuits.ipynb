```python
from qiskit import QuantumRegister, QuantumCircuit, ClassicalRegister
q = QuantumRegister(5, name='qubit')
c = ClassicalRegister(5, name='bits')
qc = QuantumCircuit(q, c)
for i in range(5):
    qc.x(q[i])
    qc.h(q[i])
qc.barrier()
for i in range(2):
    qc.y(q[i])
qc.barrier()
for i in range(2, 5):
    qc.z(q[i])
qc.draw('mpl')
```

```python
qc.draw('text')
```
