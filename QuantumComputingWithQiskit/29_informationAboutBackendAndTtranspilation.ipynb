```python
from qiskit.test.mock import FakeCambridge
backend = FakeCambridge()
```

```python
from qiskit.tools.jupyter import backend_overview
```

```python
backend
# Information about the backend.
# Topology shows how qubits can be used (eg, you can't use q1 w/ q6 or q1 w/ q7).
# Less error on a qubit makes it good for simple gate experiments, but the longer decoherence times makes it good for longer circuits.
```

```python
props = backend.properties()
```

```python
for i in range(backend.configuration().n_qubits):
    print('The U gate have lenght',props.gate_length('u3',i)*1e9,'nanosecs')
# Time the U gate takes on each qibit. Use the time of decoherence and divide it
# by the time length of a gate to know how many gates you can run before decoherence.
```

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
q = QuantumRegister(4)
c = ClassicalRegister(4)
qc = QuantumCircuit(q,c)
```

```python
qc.x(q[0])
qc.cx(q[0],q[2])
qc.barrier()
qc.x(q[1])
qc.cx(q[1],q[3])
qc.draw('mpl')
# This circuit is not possible in the backend, because q0 is not directly connected
# to q2 and q1 is not directly connected to q3.
```

```python
from qiskit import transpile
qc_compiled = transpile(qc,backend)
qc_compiled.draw('mpl')
# Transpile translates a circuit to the topology of the backend.
```

```python
from qiskit.visualization import plot_circuit_layout
plot_circuit_layout(qc_compiled,backend)
```

```python
qc_compiled.decompose().draw('mpl')
```

```python
qc_manually_compiled = transpile(qc,backend,initial_layout=[3,4,1,0])
```

```python
plot_circuit_layout(qc_manually_compiled,backend)
```

```python
qc_manually_compiled.draw('mpl')
```

```python
qc_compiled.depth() # A smaller depth means more efficiency
```

```python
qc_manually_compiled.depth()
```

```python
for i in range(4):
    qc_compiled_optimization = transpile(qc,backend,optimization_level=i)
    print('Optimization level',i)
    print('Depth',qc_compiled_optimization.depth())
    print('Gate counts',qc_compiled_optimization.count_ops())
```

```python
for i in range(4):
    qc.measure(q[i],c[i])
qc.draw('mpl')
```

```python
qc0 = transpile(qc,backend,optimization_level=0)
qc1 = transpile(qc,backend,optimization_level=1)
qc2 = transpile(qc,backend,optimization_level=2)
qc3 = transpile(qc,backend,optimization_level=3)
```

```python
qc_list = [qc0,qc1,qc2,qc3]
job = backend.run(qc_list,shots = 1024)
result = job.result()
counts = result.get_counts()
print(counts)
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts,figsize=(20,8),legend=['0','1','2','3'])
# For optimization level 2 there was a bigger chance of getting the correct state.
# All the other states are due to errors
```
