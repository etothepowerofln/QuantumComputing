```python
from qiskit import QuantumRegister, QuantumCircuit
q = QuantumRegister(2)
qc0 = QuantumCircuit(q)
qc1 = QuantumCircuit(q)
qc2 = QuantumCircuit(q)
qc3 = QuantumCircuit(q)
from qiskit.quantum_info import Statevector
qc0.h(q[0])
qc0.cx(q[0], q[1])
qc0.draw('mpl')
```

```python
phi_plus = Statevector(qc0)
phi_plus.draw('latex')
```

```python
qc1.h(q[0])
qc1.cx(q[0], q[1])
qc1.z(q[0])
qc1.draw('mpl')
```

```python
phi_minus = Statevector(qc1)
phi_minus.draw('latex')
```

```python
qc2.h(q[0])
qc2.cx(q[0], q[1])
qc2.x(q[1])
qc2.draw('mpl')
```

```python
psi_plus = Statevector(qc2)
psi_plus.draw('latex')
```

```python
qc3.h(q[0])
qc3.cx(q[0], q[1])
qc3.z(q[0])
qc3.x(q[1])
qc3.draw('mpl')
```

```python
psi_minus = Statevector(qc3)
psi_minus.draw('latex')
```

## We cannot define a state for qubit 0 and 1 separatelly in entangled qubits:

```python
from qiskit.visualization import plot_bloch_multivector
plot_bloch_multivector(qc0)
```

```python
plot_bloch_multivector(qc1)
```

```python
plot_bloch_multivector(qc2)
```

```python
plot_bloch_multivector(qc3)
```

## All the Bell's states are pure states:

```python
print(phi_plus.purity())
```

```python
print(phi_minus.purity())
```

```python
print(psi_plus.purity())
```

```python
print(psi_minus.purity())
```

## In maximal entangled states, the reduced density matrices are completelly mixed:

```python
from qiskit.quantum_info import DensityMatrix, partial_trace
rho_phi_plus = DensityMatrix(qc0)
rho_phi_plus.draw('latex', prefix='\\rho_{AB}=')
```

```python
rho_A = partial_trace(rho_phi_plus, [0])
rho_A.draw('latex', prefix='\\rho_A=')
```

```python
rho_B = partial_trace(rho_phi_plus, [1])
rho_B.draw('latex', prefix='\\rho_B=')
```

```python
print(rho_A.purity())
```

```python
print(rho_B.purity())
```
